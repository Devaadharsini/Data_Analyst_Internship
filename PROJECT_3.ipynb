```python
import pandas as pd

election_results = pd.read_excel('Election dataset 2024.xlsx')

party_alliance = pd.read_excel('Alliance.xlsx')
```

```python
print(election_results.head())

print(party_alliance.head())
```

```python
print(election_results.isnull().sum())

print(party_alliance.isnull().sum())
```

```python
election_results.fillna('Unknown', inplace=True)

party_alliance.dropna(inplace=True)
```

```python
print(election_results['Leading Party'].unique())
print(party_alliance['Party name'].unique())

# Standardize party names
election_results['Leading Party'] = election_results['Leading Party'].str.strip().str.upper()
election_results['Trailing Party'] = election_results['Trailing Party'].str.strip().str.upper()
party_alliance['Party name'] = party_alliance['Party name'].str.strip().str.upper()
```

```python
# Calculate win rate, loss rate, and percentage margin
total_records = len(election_results)
status_counts = election_results['Status'].value_counts()
result_declared_count = status_counts.get('Result Declared', 0)
uncontested_count = status_counts.get('Uncontested', 0)

win_rate = (result_declared_count / total_records) * 100
loss_rate = (uncontested_count / total_records) * 100
```

```python
election_results['Margin']
```

```python
# Convert 'Margin' column to numeric (handling errors='coerce' for non-numeric values)
election_results['Margin'] = pd.to_numeric(election_results['Margin'], errors='coerce')

# Check for non-numeric values (NaN after conversion)
non_numeric_rows = election_results[pd.isna(election_results['Margin'])]

# Print non-numeric rows (if any)
if len(non_numeric_rows) > 0:
    print("Non-numeric values found in 'Margin' column:")
    print(non_numeric_rows)
else:
    print("No non-numeric values found in 'Margin' column.")
```

```python
# Identify non-numeric values in 'Margin' column
non_numeric_mask = pd.to_numeric(election_results['Margin'], errors='coerce').isna()

# Replace NaN and non-numeric values with 0
election_results.loc[non_numeric_mask, 'Margin'] = 0
```

```python
election_results['Percentage Margin'] = (election_results['Margin'] / election_results['Margin'].sum()) * 100
```

```python
# Print results
print(f"Win Rate: {win_rate:.2f}%")
print(f"Loss Rate: {loss_rate:.2f}%")
print("Percentage Margin:")
```

```python
# Calculate number of seats won or lost by each party
party_seat_counts = election_results.groupby('Leading Party')['Status'].value_counts().unstack().fillna(0)

# Display the result
print("Number of seats won or lost by each party:")
print(party_seat_counts)
```

```python
party_seat_counts = election_results.groupby('Leading Party')['Status'].value_counts().unstack().fillna(0)

election_results = election_results.merge(party_seat_counts, left_on='Leading Party', right_index=True, how='left')
```

```python
# Calculate total number of voters
total_voters = election_results['Margin'].sum()

# Display the total number of voters
print(f"Total number of voters: {total_voters}")
```

```python
election_results.head()
```

```python
# Merge for Leading Party
merged_leading = pd.merge(election_results, party_alliance, how='left', left_on='Leading Party', right_on='Party name')
merged_leading = merged_leading.rename(columns={'Alliance': 'Leading Alliance'}).drop(columns=['Party name'])
```

```python
# Merge for Trailing Party
merged_trailing = pd.merge(merged_leading, party_alliance, how='left', left_on='Trailing Party', right_on='Party name')
merged_trailing = merged_trailing.rename(columns={'Alliance': 'Trailing Alliance'}).drop(columns=['Party name'])

# Final merged DataFrame
merged_data = merged_trailing
```

```python
merged_data.to_excel('Election_data_2024_edited_.xlsx')
```
